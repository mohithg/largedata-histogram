```python
%%time
for col in columns:
    try:
        limits = vaex_df.limits(vaex_df[col])
        bins = vaex_df.bin_edges(expression=col, limits=limits)
        histogram = vaex_df.count(expression="*", binby=vaex_df[col], edges=bins.all())
        print(histogram, bins)
    except Exception as e:
        print(e)
```
